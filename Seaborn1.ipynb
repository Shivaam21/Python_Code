```python
import seaborn as sns
```

```python
a=[1,2,3,4,5,6,7,8]
b=[5,2,11,4,7,11,7,12]
c=['yes','no','yes','no','yes','no','yes','no']
```

```python
sns.displot(b)
```

```python
sns.displot(b,bins=2)
```

```python
sns.displot(b,bins=3)
```

```python
sns.displot(b,binwidth=3)
```

```python
sns.displot(b,binwidth=2)
```

```python
sns.displot(b,binwidth=1)
```

```python
sns.displot(x=b,hue=c)
```

```python
sns.displot(x=b,y=a)
```

```python
sns.displot(x=b,col=c)
```

```python
sns.displot(x=b,y=a,hue=c)
```

```python
sns.jointplot(x=b,y=a)
```

```python
sns.relplot(x=a,y=b,hue=c)
```

```python
sns.relplot(x=a,y=b,kind="line")
```

```python
sns.relplot(x=a,y=b,kind="scatter")
```
