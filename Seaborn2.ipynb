```python
import pandas as pd
import seaborn as sns
```

```python
a=[1,2,3,4,5,6,7,8]
b=[5,2,11,4,7,11,7,12]
c=['yes','no','yes','no','yes','no','yes','no']
data={"a":a,"b":b,"c":c}
df=pd.DataFrame(data)
```

```python
df
```

```python
sns.catplot(data=df,x = c,y = a)
```

```python
sns.catplot(x=c,y=a,kind="bar",data=df)
```

```python
sns.catplot(x=c,y=a,kind="violin",data=df)
```

```python
sns.catplot(x=c,y=a,kind="point",data=df)
```

```python
sns.catplot(x=c,y=a,kind="box",data=df)
```
